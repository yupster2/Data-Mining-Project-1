### Purpose of Dataset

We chose to analyze a data set with historic daily market price for all cryptocurrencies from Kaggle. The file contained cryptocurrency trading data for 1,382 "Cryptos" with 13 attributes. The data was collected between 2013 to January 2018 for all historic open, high, low, close values for all cryptocurrencies. We were interested in the data set due to the popularity of cryptocurrencies such as Bitcoin, Ripple, Bitcoin Cash and others. Given the recent volatility in the market, we thought it would be interesting to analyze Open, High and Closing prices, while trying to determine if there is a correlation between price and relative rank of the currencies. We will run a Principal Component Analysis to determine if we can reduce our variables. We will measure the effectiveness of our algorithm by determining the percent of the variation explained by our Principal Components. We will use a "train set" of data, using the Buy Flag as our target.

### Data Understanding - Describe the meaning and type of data for each attribute in the data file

Originally, there were 13 attributes in the datset described as follows:

- slug = name of currency (created to normalize duplicate symbols);
- symbol = trading symbol of currency;
- name = common name of currency;
- date = date of trading data;
- ranknow = rank of currency;
- open = US Dollars currency price at the beginning of the day;
- high = The highest value of the currency for that day;
- low = The lowest value of the currency that day;
- close = The ending value of the currency in US dollars at the end of the day;
- volume = The value in US dollars of how many units were exchanged for the day;
- market = The market capitalization (outstanding units x closing price) in US dollars;
- close_ratio = closing price / high price for the day;
- spread = US Dollar difference between open and closing price for the day;
- Buy_Flag = variable added to determine if currency should be purchased or not; used as the target in PCA.

### Variable Characteristics / Data Quality

The data set contained 659,373 observations of 13 variables for 1,337 Crypto Tokens.  The data types of attributes are as follows: (1) Text fields: slug, symbol, name; (2) Date field: date; (3) Integer Fields: ranknow, Buy_Flag; (4) US Dollar Numeric fields: open, high, low, close, market, spread; (5) Numeric field: volume; (6) Percentage Field(Float): close_ratio

## PCA and Dimensionality Reduction

#### - Exploratory Visualization 

We will perform our PCA using only the currency prices, close ratio and spread, as other variables were unrelated to performance ...

```python
features = ['open', 'high', 'low', 'close', 'close_ratio', 'spread']
```

##### The next two lines of code separate the training set from the test set

```python
x = df.loc[:, features].values
```

```python
y = df.loc[:,['Buy_Flag']].values
```

##### The next lines of code standardize the features and run the PCA 

```python
x = StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df[['Buy_Flag']]], axis = 1)
```

Visualizing the results of the PCA shows a pretty clear delineation of Buy (0) vs Do Not Buy (1) tokens, using two Principal Components:

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0, 1]
colors = ['r', 'g', 'b']
for Buy_Flag, color in zip(targets,colors):
    indicesToKeep = finalDf['Buy_Flag'] == Buy_Flag
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

# Explained Variance
The explained variance tells us how much information (variance) can be attributed to each of the principal components. The first principal component contains 79.5% of the variance and the second principal component contains 16.8% of the variance. Together, the two components contain 96.3% of the information we need to determine Buy vs Do Not Buy!!! 

```python
pca.explained_variance_ratio_
```
